```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

x = pd.period_range(pd.datetime.now(), periods=200, freq='d')
x = x.to_timestamp().to_pydatetime()
y = np.random.randn(200, 3).cumsum(0)
```

## Plotting with Pyplot's Global State
***

```python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'
```

```python
plt.rcParams['figure.figsize'] = (8, 4)
plt.gcf().set_size_inches(8, 4)
```

#### Label and Legends

```python
plots = plt.plot(x, y)
plt.legend(plots, ('foo', 'bar', 'baz'),
           loc='best', framealpha=0.5,
           prop={'size': 'small', 'family': 'monospace'})
plt.gcf().set_size_inches(8, 4)
```

#### Titles and Axes Labels

```python
plt.title('Random trands')
plt.xlabel('Date')
plt.ylabel('Cum. sum')

plt.grid(True)
plt.figtext(0.995, 0.01,
            '\u00a9 Tramadol 0.5mg', ha='right', va='bottom')

plt.tight_layout()
plt.plot(x, y)
```

#### Saving the Plot

```python
plt.tight_layout()
plt.savefig('plot_name.png')
```
