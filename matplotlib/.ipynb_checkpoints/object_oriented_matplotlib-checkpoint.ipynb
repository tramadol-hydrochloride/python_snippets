# Object-Oriented Matplotlib
***

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

plt.style.use('seaborn-white')
x = pd.period_range(pd.datetime.now(), periods=200, freq='d')
x = x.to_timestamp().to_pydatetime()
y = np.random.randn(200, 3).cumsum(0)
```

```python
fig = plt.figure(figsize=(8, 4), dpi=200, tight_layout=True,
                 linewidth=1, edgecolor='r')
```

### Axes and Subplots
***

```python
fig = plt.figure(figsize=(8, 4))
# 0 -> 1 coordinate system, (bottom, left, top, right)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
ax.set_title('Main Axes with Insert Child Axes')
ax.plot(x, y[:, 0])
ax.set_xlabel('Date')
ax.set_ylabel('Cum. sum')

ax = fig.add_axes([0.6, 0.2, 0.3, 0.3])
ax.plot(x, y[:, 1], color='g')
ax.set_xticks([]) # remove x ticks and labels
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1,
                         sharex=True, sharey=True,
                         figsize=(8, 8))
labelled_data = zip(y.transpose(),
                    ('foo', 'bar', 'baz'), ('b', 'g', 'r'))
fig.suptitle('Three Random Treands', fontsize=16)
for i, ld in enumerate(labelled_data):
    ax = axes[i]
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.set_ylabel('Cum. sum')
    ax.legend(loc='upper left', framealpha=0.5,
              prop={'size': 'small'})
axes[-1].set_xlabel('Date')
```
